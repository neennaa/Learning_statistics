```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
study_hours=[10,15,12,20,18,8,25,14,16,30]
exam_score=[70,75,72,80,85,60,90,74,77,92]
plt.figure(figsize=(4,2))
plt.scatter(study_hours,exam_score)
plt.title("study hours vs math scores")
plt.xlabel("study hours")
plt.ylabel("exam scores")
plt.show()
```

```python
mean_study_hours=np.mean(study_hours)
mean_exam_score=np.mean(exam_score)
median_study_hours=np.median(study_hours)
median_exam_score=np.median(exam_score)
print(f"mean_study_hours:{mean_study_hours}")
print(f"mean_exam_score:{mean_exam_score}")
print(f"median_study_hours:{median_study_hours}")
print(f"median_exam_score:{median_exam_score}")
```

```python
correlation=np.corrcoef(study_hours,exam_score)[0,1]
print(f"correlation between study hours and exam score:{correlation}")
```

```python
plt.figure(figsize=(4,2))
plt.boxplot(exam_score)
plt.title("boxplot of exam scores")
plt.show()
```

```python
plt.plot(study_hours,exam_score)
plt.title("hourly exam score")
plt.xlabel("study_hours")
plt.ylabel("exam_score")
plt.show()
```
